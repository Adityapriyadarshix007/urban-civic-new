# Priority Model Experiments

This notebook experimentally evaluates supervised machine learning models
for predicting priority levels of civic complaints.

Focus:
- Train and compare priority classification models
- Analyze learning behavior and feature influence
- Study performance trade-offs between interpretability and accuracy

This notebook supports model selection decisions and does not replace
the primary rule-based or operational pipeline.

```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
```

```python
df = pd.read_csv("data/processed/features_data.csv")
df.head()
```

## 1. Problem Framing: Learning to Predict Complaint Priority

While the core system uses interpretable rule-based and tree-based models for
decision support, this experiment investigates whether machine learning models
can learn priority patterns directly from engineered spatial and categorical
features.

The goal is to study learning behavior, not to replace the primary pipeline.

```python
# Recreate priority labels (same logic as main pipeline)
def assign_priority(row):
    if row['risk_signal'] >= 0.7:
        return 2   # HIGH
    elif row['risk_signal'] >= 0.4:
        return 1   # MEDIUM
    else:
        return 0   # LOW

df['priority_label'] = df.apply(assign_priority, axis=1)
```

## 2. Experimental Setup and Learning Objective

The learning task is formulated as a multi-class classification problem where
each complaint is assigned a priority level (LOW, MEDIUM, HIGH).

Input features represent spatial density, complaint category characteristics,
and resolution status. Model performance is evaluated to understand how well
these features capture priority-related patterns.

```python
X = df[
    [
        'category_encoded',
        'status_encoded',
        'density_score',
        'category_frequency',
        'is_pending',
        'risk_signal'
    ]
]

y = df['priority_label']
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.25, random_state=42, stratify=y
)
```

## 3. Baseline Learning Model: Interpretable Decision Tree

A Decision Tree classifier is used as the baseline learning model due to its
transparent, rule-based structure. This model provides a reference point for
evaluating more complex ensemble methods.

```python
dt = DecisionTreeClassifier(
    max_depth=4,
    random_state=42
)

dt.fit(X_train, y_train)
```

## 4. Baseline Model Performance and Error Analysis

This section evaluates the Decision Tree model’s predictive performance and
examines common misclassification patterns, particularly confusion between
adjacent priority levels.

```python
dt_preds = dt.predict(X_test)

print("Decision Tree Accuracy:", accuracy_score(y_test, dt_preds))
print(classification_report(y_test, dt_preds))
```

```python
cm = confusion_matrix(y_test, dt_preds)

sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
plt.title("Decision Tree Confusion Matrix")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.show()
```

## 5. Feature Influence Analysis (Model Explainability)

Feature importance analysis is performed to understand which engineered
attributes most strongly influence priority prediction. Explainability is
critical in civic systems to ensure transparency and accountability.

```python
importances = pd.Series(
    dt.feature_importances_,
    index=X.columns
).sort_values(ascending=False)

importances.plot(kind='bar')
plt.title("Decision Tree Feature Importance")
plt.show()
```

## 6. Ensemble Learning Approach: Random Forest Classifier

A Random Forest classifier is introduced to evaluate whether ensemble learning
can improve prediction stability and accuracy by aggregating multiple decision
trees.

```python
rf = RandomForestClassifier(
    n_estimators=150,
    max_depth=6,
    random_state=42
)

rf.fit(X_train, y_train)
```

## 7. Comparative Performance Analysis: Tree vs Forest

This section compares the predictive performance of the Decision Tree and
Random Forest models to assess improvements in accuracy, robustness, and
class-wise behavior.

```python
rf_preds = rf.predict(X_test)

print("Random Forest Accuracy:", accuracy_score(y_test, rf_preds))
print(classification_report(y_test, rf_preds))
```

```python
cm_rf = confusion_matrix(y_test, rf_preds)

sns.heatmap(cm_rf, annot=True, fmt='d', cmap='Greens')
plt.title("Random Forest Confusion Matrix")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.show()
```

## 8. Comparative Feature Importance Across Models

Feature importance rankings from both models are compared to identify
consistent predictors and assess stability of learned patterns.

```python
rf_importances = pd.Series(
    rf.feature_importances_,
    index=X.columns
).sort_values(ascending=False)

rf_importances.plot(kind='bar')
plt.title("Random Forest Feature Importance")
plt.show()
```

## 9. Learning Behavior Insights and Model Selection Rationale

Experimental results indicate that ensemble learning improves predictive
stability, while simpler models retain superior interpretability. Model
selection therefore involves a trade-off between performance and transparency.

## 10. Limitations of Experimental Learning

The experimental analysis is subject to limitations such as rule-bootstrapped
labels, limited dataset size, and absence of real-time feedback. Results are
therefore indicative rather than definitive.

## 11. Role of Experimental Models in the Overall System

The experimental models validate feature design choices and provide insight
into learning behavior. The production system prioritizes interpretability,
while experimental results guide future upgrades.

## 12. Model Comparison Summary

- The Decision Tree model provides a transparent and interpretable baseline
  for priority prediction.
- The Random Forest model improves predictive accuracy and stability by
  aggregating multiple decision trees.
- Spatial density and risk-related features consistently emerge as dominant
  predictors across models.

**Conclusion:**  
Random Forest demonstrates stronger learning capability for priority prediction,
while the Decision Tree remains valuable for explainability in civic governance
contexts.
