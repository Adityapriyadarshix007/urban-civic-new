# Results Summary and Analysis
## Smart Civic Issue & Waste Management System

This notebook consolidates the final outputs of the machine learning pipeline,
including priority classification and hotspot detection.

## 1. Importing Required Libraries

This section imports libraries required for result visualization
and tabular analysis.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

import folium
```

## 2. Loading Final Result Files

The following CSV files are generated from previous ML notebooks:
- Priority classification results
- Hotspot clustering results
- Hotspot summary statistics

```python
priority_df = pd.read_csv("data/processed/priority_results.csv")
hotspot_df = pd.read_csv("data/processed/hotspot_results.csv")
hotspot_summary = pd.read_csv("data/processed/hotspot_summary.csv")

priority_df.head(), hotspot_df.head(), hotspot_summary.head()
```

## 3. Priority Level Distribution

This chart shows how civic complaints are distributed across
LOW, MEDIUM, and HIGH priority levels.

```python
priority_counts = priority_df['priority_level'].value_counts()

priority_counts.plot(kind='bar', color=['green', 'orange', 'red'])
plt.title("Priority Distribution of Civic Complaints")
plt.xlabel("Priority Level")
plt.ylabel("Number of Complaints")
plt.show()
```

## 4. Priority Counts Summary

This table summarizes the number of complaints under each
priority category.

```python
priority_counts_df = priority_counts.reset_index()
priority_counts_df.columns = ['Priority Level', 'Complaint Count']

priority_counts_df
```

## 5. Spatial Hotspot Visualization

This scatter plot visualizes DBSCAN-based hotspot clustering.
Each color represents a detected cluster, while noise points
represent isolated complaints.

```python
plt.scatter(
    hotspot_df['Longitude'],
    hotspot_df['Latitude'],
    c=hotspot_df['cluster_id'],
    cmap='tab10',
    alpha=0.7
)

plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.title("Detected Civic Complaint Hotspots")
plt.colorbar(label="Cluster ID")
plt.show()
```

## 6. Hotspot Summary Table

This table summarizes each detected hotspot in terms of:
- Number of complaints
- Average density
- Assigned hotspot risk level

```python
hotspot_summary
```

## 7. Complaint Count per Cluster

This chart shows the number of complaints present in each detected cluster.

```python
cluster_counts = hotspot_summary[
    hotspot_summary['cluster_id'] != -1
]

plt.bar(
    cluster_counts['cluster_id'].astype(str),
    cluster_counts['complaint_count']
)

plt.xlabel("Cluster ID")
plt.ylabel("Number of Complaints")
plt.title("Complaint Count per Hotspot Cluster")
plt.show()
```

## 8. Interactive Hotspot Map

This interactive map displays complaint locations colored
by detected hotspot clusters.

```python
m = folium.Map(
    location=[hotspot_df['Latitude'].mean(),
              hotspot_df['Longitude'].mean()],
    zoom_start=13
)

for _, row in hotspot_df.iterrows():
    folium.CircleMarker(
        location=[row['Latitude'], row['Longitude']],
        radius=4,
        color='red' if row['cluster_id'] != -1 else 'blue',
        fill=True,
        fill_opacity=0.6
    ).add_to(m)

m
```

```python
m.save("visualizations/hotspot_map.html")
```
