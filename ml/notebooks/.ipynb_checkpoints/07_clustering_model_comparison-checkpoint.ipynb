## DBSCAN vs KMeans for Hotspot Detection

This experiment compares density-based clustering (DBSCAN) with centroid-based
clustering (KMeans) for detecting civic complaint hotspots.

The objective is to study how different clustering assumptions affect
hotspot identification in spatial civic data.

## Experimental Setup

Both clustering algorithms are applied to the same spatial coordinates
(latitude and longitude). No categorical or engineered features are used
to avoid bias in spatial grouping.

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import DBSCAN, KMeans

df = pd.read_csv("data/processed/features_data.csv")
coords = df[['Latitude', 'Longitude']].values
```

## DBSCAN Clustering

DBSCAN groups points based on local density and automatically identifies
noise points. It does not require a predefined number of clusters.


```python
dbscan = DBSCAN(eps=0.002, min_samples=5)
df['dbscan_cluster'] = dbscan.fit_predict(coords)
```

```python
plt.scatter(
    df['Longitude'],
    df['Latitude'],
    c=df['dbscan_cluster'],
    cmap='tab10',
    alpha=0.6
)
plt.title("DBSCAN Hotspot Detection")
plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.colorbar(label="Cluster ID")
plt.show()
```

## KMeans Clustering

KMeans partitions data into a predefined number of clusters by minimizing
distance to cluster centroids. It assumes spherical cluster shapes.

```python
kmeans = KMeans(n_clusters=4, random_state=42)
df['kmeans_cluster'] = kmeans.fit_predict(coords)
```

```python
plt.scatter(
    df['Longitude'],
    df['Latitude'],
    c=df['kmeans_cluster'],
    cmap='tab10',
    alpha=0.6
)
plt.title("KMeans Clustering Output")
plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.colorbar(label="Cluster ID")
plt.show()
```

```python
print("DBSCAN cluster distribution:")
print(df['dbscan_cluster'].value_counts())

print("\nKMeans cluster distribution:")
print(df['kmeans_cluster'].value_counts())
```

```python
fig, axes = plt.subplots(1, 2, figsize=(14, 6))

axes[0].scatter(
    df['Longitude'], df['Latitude'],
    c=df['dbscan_cluster'], cmap='tab10', alpha=0.6
)
axes[0].set_title("DBSCAN")

axes[1].scatter(
    df['Longitude'], df['Latitude'],
    c=df['kmeans_cluster'], cmap='tab10', alpha=0.6
)
axes[1].set_title("KMeans")

plt.show()
```

## Comparative Analysis and Insights

DBSCAN identifies organically formed hotspots and isolates sparse complaints
as noise, making it suitable for real-world civic data with uneven density.

KMeans forces all complaints into predefined clusters, which may lead to
artificial grouping and reduced interpretability in hotspot detection.

These results validate the selection of DBSCAN for the primary hotspot
detection pipeline.

## Model Selection Decision

Based on experimental comparison, DBSCAN is selected for hotspot detection
due to its ability to handle irregular spatial distributions and noise.
KMeans is retained only for comparative analysis.
