## Anomaly Detection in Civic Complaints

This experiment applies unsupervised anomaly detection to identify
unusual civic complaints based on spatial density and complaint patterns.

The objective is to highlight cases that may require special attention
or manual inspection by authorities.


## Feature Space for Anomaly Detection

Anomaly detection is performed using a subset of engineered features that
capture spatial density and complaint frequency characteristics.

```python
import pandas as pd
from sklearn.ensemble import IsolationForest
import matplotlib.pyplot as plt

df = pd.read_csv("data/processed/features_data.csv")
```

```python
X_anomaly = df[
    [
        'density_score',
        'category_frequency',
        'is_pending'
    ]
]
```

## Isolation Forest Model

Isolation Forest isolates anomalies by randomly partitioning data.
Points that require fewer partitions are considered anomalous.

```python
iso = IsolationForest(
    n_estimators=150,
    contamination=0.08,
    random_state=42
)

df['anomaly_flag'] = iso.fit_predict(X_anomaly)
```

```python
df['anomaly_flag'].value_counts()
```

## Spatial Distribution of Detected Anomalies

Anomalous complaints are visualized to examine whether they form
spatial patterns or appear as isolated outliers.

```python
plt.scatter(
    df['Longitude'],
    df['Latitude'],
    c=df['anomaly_flag'],
    cmap='coolwarm',
    alpha=0.6
)

plt.title("Spatial Visualization of Anomalous Complaints")
plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.colorbar(label="Anomaly Flag")
plt.show()
```

## Characteristics of Anomalous Complaints

This section examines how anomalous complaints differ from normal cases
in terms of density and category frequency.

```python
df.groupby('anomaly_flag')[
    ['density_score', 'category_frequency', 'is_pending']
].mean()
```

## Interpretation and Civic Relevance

Detected anomalies may represent:
- Highly congested civic problem zones
- Rare but severe complaint types
- Long-pending unresolved cases

Such complaints can be flagged for priority inspection or escalation.

## Limitations of Anomaly Detection

Anomaly detection does not imply error or wrongdoing.
Results are sensitive to feature selection and contamination assumptions
and should be used as a decision-support tool.

## Summary of Experimental Findings

Isolation Forest successfully identified a small subset of unusual
complaints that differ significantly from the majority.

This experiment demonstrates the potential of unsupervised learning
to enhance civic monitoring and decision-making.
